```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("saudi_projects_updated.csv")
df = df[['sectors','sector_budgets','sector_num_projects','type_project','budget_project','startday_project','enddate_project',
        'project_area','region_project','status_project']]
df.head()
```

```python
# Change the date columns to date format
df['startday_project'] = pd.to_datetime(df['startday_project'],format='%d-%m-%Y')
df['enddate_project'] = pd.to_datetime(df['enddate_project'],format='%d-%m-%Y')
df
```

```python
# Add duration column
df['duration_project'] = df['enddate_project'] - df['startday_project'] 
df
```

```python
df['duration_project'] = df['duration_project'].astype('timedelta64[D]')
```

```python
# Add month and year columns
df['start_year'] = pd.DatetimeIndex(df['startday_project']).year
df['start_month'] = pd.DatetimeIndex(df['startday_project']).month
df['end_year'] = pd.DatetimeIndex(df['enddate_project']).year
df['end_month'] = pd.DatetimeIndex(df['enddate_project']).month
df
```

```python
# Save the Data set with the colums needed for the models
df = df[['sectors','sector_budgets','type_project','budget_project','startday_project','start_year','start_month',
         'enddate_project','end_year','end_month','duration_project','project_area','region_project','status_project']]
df.head()
```

```python
df.to_csv(r"saudi_projects_v02.csv")
```
